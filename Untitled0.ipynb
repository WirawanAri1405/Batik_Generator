```python
# @title 2. Download Dataset & EDA Batik
import kagglehub
import shutil
import glob
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm

# --- 1. DOWNLOAD & PREPROCESSING ---

# Download Dataset
path = kagglehub.dataset_download("dionisiusdh/indonesian-batik-motifs")

# Setup folder output
data_dir = "/content/batik_data"
if os.path.exists(data_dir): shutil.rmtree(data_dir)
os.makedirs(data_dir, exist_ok=True)

print("Memproses gambar (Resize & Convert RGB)...")
files = glob.glob(os.path.join(path, '**', '*.jpg'), recursive=True) + \
        glob.glob(os.path.join(path, '**', '*.png'), recursive=True)

target_size = 512
processed_count = 0

for idx, f in enumerate(tqdm(files)):
    try:
        # Load, convert ke RGB (hilangkan alpha channel jika png), dan resize
        img = Image.open(f).convert('RGB').resize((target_size, target_size))
        save_path = f"{data_dir}/{idx}.jpg"
        img.save(save_path, quality=95)
        processed_count += 1
    except Exception as e:
        print(f"Gagal memproses {f}: {e}")

print(f"\n✅ Siap! {processed_count} gambar tersimpan di {data_dir}")

# --- 2. EXPLORATORY DATA ANALYSIS (EDA) ---

print("\n=== MEMULAI EDA (Exploratory Data Analysis) ===")
final_files = glob.glob(os.path.join(data_dir, '*.jpg'))

# A. Tampilkan Sampel Grid Gambar
def show_grid(image_paths, n=16):
    plt.figure(figsize=(10, 10))
    # Ambil sampel acak
    sample_imgs = random.sample(image_paths, min(len(image_paths), n))

    for i, img_path in enumerate(sample_imgs):
        plt.subplot(4, 4, i + 1)
        img = Image.open(img_path)
        plt.imshow(img)
        plt.axis('off') # Hilangkan axis agar rapi
        plt.title(f"Img {os.path.basename(img_path)}")

    plt.suptitle(f"Sampel 16 Gambar Batik ({target_size}x{target_size})", fontsize=16)
    plt.tight_layout()
    plt.show()

if len(final_files) > 0:
    show_grid(final_files)
else:
    print("Tidak ada gambar untuk ditampilkan.")

# B. Analisis Distribusi Warna (Histogram RGB)
# Kita ambil sampel 100 gambar saja agar proses cepat
def plot_color_dist(image_paths, sample_size=100):
    print("\nMenganalisis distribusi warna...")
    sample_paths = random.sample(image_paths, min(len(image_paths), sample_size))

    red_vals = []
    green_vals = []
    blue_vals = []

    for p in tqdm(sample_paths):
        img = np.array(Image.open(p))
        # Flatten array per channel
        red_vals.extend(img[:, :, 0].flatten())
        green_vals.extend(img[:, :, 1].flatten())
        blue_vals.extend(img[:, :, 2].flatten())

    plt.figure(figsize=(12, 5))
    plt.hist(red_vals, bins=50, color='red', alpha=0.5, label='Red Channel')
    plt.hist(green_vals, bins=50, color='green', alpha=0.5, label='Green Channel')
    plt.hist(blue_vals, bins=50, color='blue', alpha=0.5, label='Blue Channel')
    plt.title('Distribusi Warna RGB (Sampel 100 Gambar)')
    plt.xlabel('Intensitas Piksel (0-255)')
    plt.ylabel('Frekuensi')
    plt.legend()
    plt.show()

if len(final_files) > 0:
    plot_color_dist(final_files)
```

```python
# @title 3. Advanced EDA: Blur, Brightness & Duplicates
!pip install imagehash -q

import cv2
import numpy as np
import imagehash
import matplotlib.pyplot as plt
from PIL import Image
import os
import glob
from tqdm import tqdm

data_dir = "/content/batik_data"
files = glob.glob(os.path.join(data_dir, '*.jpg'))

# List untuk menyimpan metrik
blur_scores = []
brightness_scores = []
hashes = {}
duplicates = []

print("Menganalisis kualitas gambar (Blur, Brightness, & Duplicates)...")

for f in tqdm(files):
    # Baca gambar dengan OpenCV untuk hitungan matematis
    img_cv = cv2.imread(f)
    img_pil = Image.open(f)

    # 1. Cek BLUR menggunakan Laplacian Variance
    # Semakin rendah nilainya, semakin blur gambarnya.
    gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
    variance = cv2.Laplacian(gray, cv2.CV_64F).var()
    blur_scores.append(variance)

    # 2. Cek BRIGHTNESS rata-rata
    avg_brightness = np.mean(gray)
    brightness_scores.append(avg_brightness)

    # 3. Cek DUPLIKAT menggunakan Perceptual Hashing (pHash)
    # pHash tetap mendeteksi duplikat meskipun gambar sedikit diedit/resize
    h = imagehash.phash(img_pil)
    if h in hashes:
        duplicates.append(f)
    else:
        hashes[h] = f

# --- VISUALISASI HASIL ---

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Plot Blur
ax[0].hist(blur_scores, bins=50, color='purple', alpha=0.7)
ax[0].set_title('Distribusi Ketajaman (Laplacian Variance)')
ax[0].set_xlabel('Score (Makin kecil = Makin Blur)')
ax[0].set_ylabel('Jumlah Gambar')
# Garis batas wajar (biasanya < 100 dianggap agak blur)
ax[0].axvline(100, color='r', linestyle='dashed', linewidth=1, label='Threshold Blur (100)')
ax[0].legend()

# Plot Brightness
ax[1].hist(brightness_scores, bins=50, color='orange', alpha=0.7)
ax[1].set_title('Distribusi Kecerahan (Mean Pixel Intensity)')
ax[1].set_xlabel('0 (Gelap) - 255 (Terang)')
ax[1].axvline(np.mean(brightness_scores), color='k', linestyle='dashed', linewidth=1, label='Rata-rata')
ax[1].legend()

plt.show()

# --- LAPORAN TEKS ---
print(f"\n=== LAPORAN KUALITAS DATA ===")
print(f"Total Gambar Awal   : {len(files)}")
print(f"Rata-rata Ketajaman : {np.mean(blur_scores):.2f}")
print(f"Gambar Potensi Blur : {sum(1 for x in blur_scores if x < 100)} (Score < 100)")
print(f"Gambar Potensi Duplikat : {len(duplicates)}")

if len(duplicates) > 0:
    print(f"Contoh file duplikat: {duplicates[:3]}")
    print("Saran: Sebaiknya hapus duplikat agar model tidak overfitting.")

# Opsi: Menampilkan gambar paling blur untuk dicek
print("\n--- Cek 5 Gambar Paling Blur ---")
sorted_blur_indices = np.argsort(blur_scores)
plt.figure(figsize=(15, 3))
for i in range(5):
    idx = sorted_blur_indices[i]
    path = files[idx]
    score = blur_scores[idx]

    plt.subplot(1, 5, i+1)
    plt.imshow(Image.open(path))
    plt.axis('off')
    plt.title(f"Score: {score:.1f}")
plt.show()
```

```python
# @title 4. Inspeksi Visual Gambar Bermasalah
import matplotlib.pyplot as plt
import math
from PIL import Image

def show_problem_images(image_paths, title, color_title='red'):
    n = len(image_paths)
    if n == 0:
        print(f"✅ Tidak ada {title} yang ditemukan.")
        return

    cols = 8  # Jumlah kolom per baris
    rows = math.ceil(n / cols)

    plt.figure(figsize=(20, 2.5 * rows))
    plt.suptitle(f"{title} (Total: {n})", fontsize=20, color=color_title, y=1.02)

    for i, img_path in enumerate(image_paths):
        plt.subplot(rows, cols, i + 1)
        try:
            img = Image.open(img_path)
            plt.imshow(img)
            plt.axis('off')
            # Tampilkan nama file agar mudah dilacak
            plt.title(os.path.basename(img_path), fontsize=8)
        except:
            pass

    plt.tight_layout()
    plt.show()

# --- 1. TAMPILKAN GAMBAR BLUR (< 100) ---
# Ambil path gambar yang score-nya di bawah 100
blurry_images = [files[i] for i, score in enumerate(blur_scores) if score < 100]
show_problem_images(blurry_images, "Gambar Terdeteksi Blur (Score < 100)", color_title='purple')

# --- 2. TAMPILKAN DUPLIKAT ---
# Variabel 'duplicates' sudah berisi list path dari cell sebelumnya
show_problem_images(duplicates, "Gambar Terdeteksi Duplikat", color_title='orange')
```

```python
# @title 6. Auto-Remove Human/Models using YOLOv8
!pip install ultralytics -q

from ultralytics import YOLO
import shutil
import os
import glob
from tqdm import tqdm

# Setup Folder
source_dir = "/content/batik_data"
clean_dir = "/content/batik_data_clean"      # Folder untuk batik murni
rejected_dir = "/content/batik_with_humans"  # Folder untuk yang ada orangnya

if os.path.exists(clean_dir): shutil.rmtree(clean_dir)
if os.path.exists(rejected_dir): shutil.rmtree(rejected_dir)
os.makedirs(clean_dir, exist_ok=True)
os.makedirs(rejected_dir, exist_ok=True)

# Load Model YOLOv8 Nano (Versi paling ringan & cepat)
print("Loading model YOLOv8...")
model = YOLO('yolov8n.pt')

files = glob.glob(os.path.join(source_dir, '*.jpg'))
print(f"Menganalisis {len(files)} gambar untuk mendeteksi manusia...")

count_human = 0
count_clean = 0

# Loop gambar
for f in tqdm(files):
    filename = os.path.basename(f)

    # Run inference (matikan verbose biar log gak penuh)
    results = model(f, verbose=False, conf=0.5) # conf=0.5 artinya yakin 50% itu orang

    # Cek apakah ada class 'person' (Class ID 0 di COCO dataset adalah Person)
    # results[0].boxes.cls berisi list ID objek yang terdeteksi
    detected_classes = results[0].boxes.cls.cpu().numpy()

    if 0 in detected_classes: # 0 = Person
        shutil.copy(f, os.path.join(rejected_dir, filename))
        count_human += 1
    else:
        shutil.copy(f, os.path.join(clean_dir, filename))
        count_clean += 1

print("\n=== HASIL FILTERISASI MANUSIA ===")
print(f"✅ Batik Bersih (Disimpan di {clean_dir}): {count_clean}")
print(f"❌ Terdeteksi Manusia (Disimpan di {rejected_dir}): {count_human}")
```

```python
# @title 7. Cek Visual Hasil Filter YOLO
import matplotlib.pyplot as plt
import glob
import random
from PIL import Image
import os
import math

def show_sample_gallery(folder_path, title, n=10):
    files = glob.glob(os.path.join(folder_path, '*.jpg'))
    if len(files) == 0:
        print(f"Folder {title} kosong.")
        return

    sample_files = random.sample(files, min(len(files), n))

    cols = 5
    rows = math.ceil(len(sample_files) / cols)

    plt.figure(figsize=(15, 3 * rows))
    plt.suptitle(title, fontsize=16, y=1.02)

    for i, f in enumerate(sample_files):
        plt.subplot(rows, cols, i+1)
        plt.imshow(Image.open(f))
        plt.axis('off')
        plt.title(os.path.basename(f), fontsize=8)
    plt.tight_layout()
    plt.show()

# Tampilkan sampel yang DIBUANG (seharusnya isinya orang/model)
show_sample_gallery("/content/batik_with_humans", "Sampel DIBUANG (Terdeteksi Manusia)", n=100)

# Tampilkan sampel yang DISIMPAN (seharusnya hanya kain/pola)
show_sample_gallery("/content/batik_data_clean", "Sampel BERSIH (Siap Training)", n=1000)
```
