```python
import os
import glob
from IPython.display import Image, display

# Path sesuai log yang kamu berikan
output_folder = '/content/results/batik_parang-generated-8'

# Mencari semua file gambar (jpg/png) di folder tersebut
image_files = glob.glob(os.path.join(output_folder, '*'))

print(f"Ditemukan {len(image_files)} gambar.")

# Menampilkan setiap gambar
for img_path in image_files:
    if img_path.lower().endswith(('.png', '.jpg', '.jpeg')):
        print(f"Menampilkan: {os.path.basename(img_path)}")
        display(Image(filename=img_path, width=256)) # width bisa diatur
```

```python
import os
import glob
from IPython.display import Image, display

# Lokasi folder hasil (sesuaikan jika path berubah seperti sebelumnya)
# Biasanya formatnya: {results_dir}/{name}-interpolation-{checkpoint}/
output_folder_pattern = '/content/results/*'

# Cari folder terbaru
folders = glob.glob(output_folder_pattern)
if folders:
    latest_folder = max(folders, key=os.path.getmtime)
    print(f"Folder ditemukan: {latest_folder}")

    # Cari file GIF di dalamnya
    gif_files = glob.glob(os.path.join(latest_folder, '*.gif'))

    if gif_files:
        for gif in gif_files:
            print(f"Menampilkan: {os.path.basename(gif)}")
            display(Image(filename=gif))
    else:
        print("File GIF belum ketemu, coba cek folder secara manual.")
else:
    print("Folder hasil interpolasi belum ditemukan. Pastikan proses generate selesai.")
```
